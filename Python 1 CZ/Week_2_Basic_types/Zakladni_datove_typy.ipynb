```python
# Assigning values to variables
x = 5
y = 2.5
z = 3 + 4j

# Basic arithmetic operations
a = x + y    # Addition
b = x - y    # Subtraction
c = x * y    # Multiplication
d = x / y    # Division
e = x ** y   # Exponentiation
f = x % y    # Modulo

# Displaying the results
print("a =", a)
print("b =", b)
print("c =", c)
print("d =", d)
print("e =", e)
print("f =", f)

# Complex number arithmetic
g = z + 2    # Addition
h = z * 3    # Multiplication

# Displaying the results
print("g =", g)
print("h =", h)
```

```python
result = 10 + 5 * 2
print(result)   # Output: 20
```

Pokud bychom chtěli změnit pořadí operací, mohli bychom výrazy seskupit pomocí závorek. Například:

```python
result = (10 + 5) * 2
print(result)   # Output: 30
```

## Řetězce

Řetězec je posloupnost znaků uzavřená v uvozovkách (jednoduchých nebo dvojitých). Zde je několik klíčových funkcí a operací, které byste měli znát při práci s řetězci v jazyce Python:

 - **Indexování**: K jednotlivým znakům řetězce můžete přistupovat pomocí indexování. První znak řetězce má index $0$, druhý znak má index `1` a tak dále. K přístupu ke znakům od konce řetězce můžete použít také záporné indexování, přičemž poslední znak má index `-1`, předposlední znak má index `-2` atd.

 - **Slicing**: Pomocí řezání můžete extrahovat část řetězce. Krájení se provádí pomocí hranatých závorek `[]` s počátečním indexem (včetně), koncovým indexem (výlučně) a velikostí kroku (nepovinně) oddělenými dvojtečkami `:`. Například `my_ string[2:5]` by extrahovalo znaky od indexu `2` do indexu `4` řetězce my_string.

 - **Konkatenace**: Pomocí operátoru + můžete spojit dva nebo více řetězců. Například `"Hello" + " " + "world"` by vytvořilo řetězec `"Hello world"`.

 - **Zopakování**: Pomocí operátoru `*` můžete řetězec určitý početkrát zopakovat. Například `"Hello" * 3` by vytvořilo řetězec `"HelloHelloHello"`.

 - **Délka**: Délku řetězce (tj. počet znaků v řetězci) můžete zjistit pomocí funkce `len()`. Například funkce `len("Hello")` by vrátila hodnotu `5`.

 - **Formátování**: Řetězec můžete formátovat pomocí zástupných znaků a metody `format()`. Zástupné znaky jsou označeny kudrnatými závorkami `{}` a lze je nahradit hodnotami pomocí metody `format()`. Například `"Hodnota x je {} a hodnota y je {}".format(x, y)` by vytvořilo řetězec s hodnotami x a y vloženými do zástupných znaků.

```python
# Indexing and slicing
my_string = "Hello world"
print(my_string[0])       # Output: 'H'
print(my_string[-1])      # Output: 'd'
print(my_string[3:7])     # Output: 'lo w'
print(my_string[:5])      # Output: 'Hello'
print(my_string[6:])      # Output: 'world'

# Concatenation and repetition
greeting = "Hello"
name = "Jan"
message = greeting + " " + name + "!"
print(message)            # Output: 'Hello Jan!'
print(name * 3)           # Output: 'JanJanJan'

# Length
print(len(my_string))     # Output: 11

# Formatting
x = 5
y = 10
formatted_string = "The value of x is {} and the value of y is {}".format(x, y)
print(formatted_string)   # Output: 'The value of x is 5 and the value of y is 10'
```

## Seznamy

Seznam je kolekce položek uzavřených v hranatých závorkách `[]` a oddělených čárkami. Zde je několik klíčových funkcí a operací, které byste měli znát při práci se seznamy v jazyce Python:

 - **Indexování**: K jednotlivým prvkům seznamu můžete přistupovat pomocí indexování. První prvek seznamu má index `0`, druhý prvek má index `1` atd. K přístupu k prvkům od konce seznamu můžete použít také záporné indexování, přičemž poslední prvek má index `-1`, předposlední prvek má index `-2` atd.

- **Slicing**: Pomocí řezání můžete získat část seznamu. Krájení se provádí pomocí hranatých závorek `[]` s počátečním indexem (včetně), koncovým indexem (bez) a velikostí kroku (nepovinně) oddělenými dvojtečkami `:`. Například `my_list[2:5]` by extrahovalo prvky od indexu `2` do indexu `4` seznamu `my_list`.

- **Modifikace**: Seznamy jsou proměnlivé, což znamená, že je můžete upravovat přidáváním, odebíráním nebo změnou prvků. Můžete přidat prvek na konec seznamu pomocí metody `append()`, vložit prvek na určitý index pomocí metody `insert()` nebo odebrat prvek podle jeho hodnoty pomocí metody `remove()`.

- **Délka**: Délku seznamu (tj. počet prvků v seznamu) můžete zjistit pomocí funkce `len()`. Například funkce `len([1, 2, 3])` vrátí hodnotu `3`.

- **Iterace**: Prvky seznamu můžete iterovat pomocí cyklu `for`.

```python
# Indexing and slicing
my_list = [1, 2, 3, 4, 5]
print(my_list[0])       # Output: 1
print(my_list[-1])      # Output: 5
print(my_list[2:4])     # Output: [3, 4]
print(my_list[:3])      # Output: [1, 2, 3]
print(my_list[3:])      # Output: [4, 5]

# Modifying
my_list.append(6)
print(my_list)          # Output: [1, 2, 3, 4, 5, 6]

my_list.insert(2, "Hello")
print(my_list)          # Output: [1, 2, 'Hello', 3, 4, 5, 6]

my_list.remove(3)
print(my_list)          # Output: [1, 2, 'Hello', 4, 5, 6]

# Length
print(len(my_list))     # Output: 6

# Iterating
for element in my_list:
    print(element)
```

## Slovníky

Slovník je kolekce dvojic klíč-hodnota uzavřená do složených závorek `{}` a oddělená čárkami. Zde je několik klíčových funkcí a operací, které byste měli znát při práci se slovníky v jazyce Python:

 - **Přístup k hodnotám**: K hodnotě položky slovníku můžete přistupovat zadáním jejího klíče uvnitř hranatých závorek `[]`. Například `my_dict["key"]` vrátí hodnotu přiřazenou klíči `"key"` ve slovníku `my_dict`.

 - **Modifikace hodnot**: Slovníky jsou mutovatelné, což znamená, že jejich hodnoty můžete měnit přiřazením nových hodnot konkrétním klíčům. Například `my_dict["key"] = "new value"` by nastavilo hodnotu klíče `"key"` ve slovníku `my_dict` na `"new value"`.

 - **Přidávání položek**: Novou položku do slovníku přidáte zadáním nové dvojice klíč-hodnota pomocí syntaxe `my_dict[new_key] = new_value`. Například `my_dict["new_key"] = "new_value"` přidá novou položku do slovníku `my_dict`.

 - **Odstranění položek**: Odstranění položky ze slovníku lze provést příkazem del, za kterým následuje klíč položky, kterou chcete odstranit. Například příkaz `del my_dict["key"]` odstraní ze slovníku `my_dict` položku s klíčem `"key"`.

 - **Délka**: Počet položek ve slovníku můžete zjistit pomocí funkce `len()`. Například funkce `len(my_dict)` by vrátila počet položek ve slovníku `my_dict`.

 - **Iterace**: Klíče nebo hodnoty slovníku můžete iterovat pomocí cyklu `for`.

```python
# Accessing values
my_dict = {"name": "Jana", "age": 30, "gender": "female"}
print(my_dict["name"])   # Output: 'Jana'
print(my_dict["age"])    # Output: 30

# Modifying values
my_dict["age"] = 40
print(my_dict["age"])    # Output: 40

# Adding items
my_dict["occupation"] = "engineer"
print(my_dict)           # Output: {'name': 'Jana', 'age': 40, 'gender': 'female', 'occupation': 'engineer'}

# Removing items
del my_dict["gender"]
print(my_dict)           # Output: {'name': 'Jana', 'age': 40, 'occupation': 'engineer'}

# Length
print(len(my_dict))      # Output: 3

# Iterating
for key in my_dict:
    print(key, my_dict[key])
```

## Hnízdění
Různé datové typy jazyka Python můžete vnořovat do sebe. Zde je příklad, jak můžete vnořit seznam do slovníku a slovník do seznamu:

```python
# Nesting a list inside a dictionary
my_dict = {"name": "Jan", "age": 20, "grades": ['A', 'B', 'C', 'D']}
print(my_dict["grades"][0])   # Output: "A"

# Nesting a dictionary inside a list
my_list = [{"name": "Jan", "age": 30}, {"name": "Jana", "age": 25}]
print(my_list[0]["name"])     # Output: "Jan"
```
