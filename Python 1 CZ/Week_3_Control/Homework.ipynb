# Homework

Create a simple shopping list application that allows users to add and remove items from a shopping list. The program should have the following features:

 - Display the current shopping list to the user, including the name and quantity of each item.

 - Allow the user to add items to the shopping list by entering the name and quantity of each item.

 - Allow the user to remove items from the shopping list by entering the name of the item.

 - Allow the user to update the quantity of items in the shopping list.

 - Allow the user to view the total cost of all items in the shopping list, based on a fixed price for each item.

To implement this project, you can use a dictionary to store the shopping list items, where the keys are the names of the items and the values are the quantities of each item. You can use a loop to repeatedly prompt the user for input and carry out the desired operations, such as adding or removing items from the shopping list.

```python
# Initialize an empty shopping list
shopping_list = {}

# Define the prices of the items
item_prices = {"apple": 0.5, "banana": 0.25, "bread": 2.0, "milk": 1.0}

# Define a function to calculate the total cost of the shopping list
def calculate_total(shopping_list, item_prices):
    total = 0
    for item in shopping_list:
        total += item_prices[item] * shopping_list[item]
    return total

# Main loop
while True:
    # Display the current shopping list
    print("Shopping List:")
    for item in shopping_list:
        print(item, shopping_list[item])

    # Prompt the user for input
    action = input("Enter an action (add, remove, update, total, or quit): ")

    # Add an item to the shopping list
    if action == "add":
        item = input("Enter the name of the item: ")
        quantity = int(input("Enter the quantity: "))
        shopping_list[item] = shopping_list.get(item, 0) + quantity

    # Remove an item from the shopping list
    elif action == "remove":
        item = input("Enter the name of the item: ")
        if item in shopping_list:
            del shopping_list[item]

    # Update the quantity of an item in the shopping list
    elif action == "update":
        item = input("Enter the name of the item: ")
        if item in shopping_list:
            quantity = int(input("Enter the new quantity: "))
            shopping_list[item] = quantity

    # Calculate the total cost of the shopping list
    elif action == "total":
        total_cost = calculate_total(shopping_list, item_prices)
        print("Total cost: $", total_cost)

    # Quit the program
    elif action == "quit":
        break

    # Invalid action
    else:
        print("Invalid action. Please try again.")
```
