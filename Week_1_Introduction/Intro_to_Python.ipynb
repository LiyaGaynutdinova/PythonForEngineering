## Numbers
In this example, we first assign some values to variables x, y, and z representing an integer, a floating-point number, and a complex number, respectively. We then perform some basic arithmetic operations using these variables, such as addition, subtraction, multiplication, division, exponentiation, and modulo, and assign the results to new variables a through f. Finally, we perform some complex number arithmetic using variable z and display the results.

When you run this code, you should see the results of each operation displayed on the screen.

```python
# Assigning values to variables
x = 5
y = 2.5
z = 3 + 4j

# Basic arithmetic operations
a = x + y    # Addition
b = x - y    # Subtraction
c = x * y    # Multiplication
d = x / y    # Division
e = x ** y   # Exponentiation
f = x % y    # Modulo

# Displaying the results
print("a =", a)
print("b =", b)
print("c =", c)
print("d =", d)
print("e =", e)
print("f =", f)

# Complex number arithmetic
g = z + 2    # Addition
h = z * 3    # Multiplication

# Displaying the results
print("g =", g)
print("h =", h)
```

The order of operations is as follows:

 - Parentheses: Any expressions inside parentheses should be evaluated first.

 - Exponents: Any expressions involving exponents (such as **) should be evaluated next, from left to right.

 - Multiplication and Division: Multiplication (*) and division (/) should be evaluated next, from left to right.

 - Addition and Subtraction: Addition (+) and subtraction (-) should be evaluated last, from left to right.

In this example, the expression 10 + 5 * 2 is evaluated according to the order of operations.

```python
result = 10 + 5 * 2
print(result)   # Output: 20
```

If we wanted to change the order of operations, we could use parentheses to group expressions together. For example:

```python
result = (10 + 5) * 2
print(result)   # Output: 30
```

## Strings

A string is a sequence of characters enclosed in quotation marks (either single or double). Here are some key features and operations you should be familiar with when working with strings in Python:

 - **Indexing**: You can access individual characters of a string using indexing. The first character of the string has index $0$, the second character has index `1` and so on. You can also use negative indexing to access characters from the end of the string, with the last character having index `-1`, the second-last character having index `-2`, and so on.

 - **Slicing**: You can extract a portion of a string using slicing. Slicing is done using square brackets `[]` with the start index (inclusive), end index (exclusive), and step size (optional) separated by colons `:`. For example, `my_ string[2:5]` would extract the characters from index `2` to index `4` of the string my_string.

 - **Concatenation**: You can concatenate two or more strings using the + operator. For example, `"Hello" + " " + "world"` would produce the string `"Hello world"`.

 - **Repetition**: You can repeat a string a certain number of times using the `*` operator. For example, `"Hello" * 3` would produce the string `"HelloHelloHello"`.

 - **Length**: You can find the length of a string (i.e., the number of characters in the string) using the `len()` function. For example, `len("Hello")` would return the value `5`.

 - **Formatting**: You can format a string using placeholders and the `format()` method. Placeholders are denoted by curly braces `{}` and can be replaced with values using the `format()` method. For example, `"The value of x is {} and the value of y is {}".format(x, y)` would produce a string with the values of x and y inserted into the placeholders.

```python
# Indexing and slicing
my_string = "Hello world"
print(my_string[0])       # Output: 'H'
print(my_string[-1])      # Output: 'd'
print(my_string[3:7])     # Output: 'lo w'
print(my_string[:5])      # Output: 'Hello'
print(my_string[6:])      # Output: 'world'

# Concatenation and repetition
greeting = "Hello"
name = "Jan"
message = greeting + " " + name + "!"
print(message)            # Output: 'Hello Jan!'
print(name * 3)           # Output: 'JanJanJan'

# Length
print(len(my_string))     # Output: 11

# Formatting
x = 5
y = 10
formatted_string = "The value of x is {} and the value of y is {}".format(x, y)
print(formatted_string)   # Output: 'The value of x is 5 and the value of y is 10'
```

## Lists

A list is a collection of items enclosed in square brackets `[]` and separated by commas. Here are some key features and operations you should be familiar with when working with lists in Python:

 - **Indexing**: You can access individual elements of a list using indexing. The first element of the list has index `0`, the second element has index `1`, and so on. You can also use negative indexing to access elements from the end of the list, with the last element having index `-1`, the second-last element having index `-2`, and so on.

- **Slicing**: You can extract a portion of a list using slicing. Slicing is done using square brackets `[]` with the start index (inclusive), end index (exclusive), and step size (optional) separated by colons `:`. For example, `my_list[2:5]` would extract the elements from index `2` to index `4` of the list `my_list`.

- **Modifying**: Lists are mutable, meaning you can modify them by adding, removing, or changing elements. You can append an element to the end of a list using the `append()` method, insert an element at a specific index using the `insert()` method, or remove an element by its value using the `remove()` method.

- **Length**: You can find the length of a list (i.e., the number of elements in the list) using the `len()` function. For example, `len([1, 2, 3])` would return the value `3`.

- **Iterating**: You can iterate over the elements of a list using a `for` loop. 

```python
# Indexing and slicing
my_list = [1, 2, 3, 4, 5]
print(my_list[0])       # Output: 1
print(my_list[-1])      # Output: 5
print(my_list[2:4])     # Output: [3, 4]
print(my_list[:3])      # Output: [1, 2, 3]
print(my_list[3:])      # Output: [4, 5]

# Modifying
my_list.append(6)
print(my_list)          # Output: [1, 2, 3, 4, 5, 6]

my_list.insert(2, "Hello")
print(my_list)          # Output: [1, 2, 'Hello', 3, 4, 5, 6]

my_list.remove(3)
print(my_list)          # Output: [1, 2, 'Hello', 4, 5, 6]

# Length
print(len(my_list))     # Output: 6

# Iterating
for element in my_list:
    print(element)
```

## Dictionaries

A dictionary is a collection of key-value pairs enclosed in curly braces `{}` and separated by commas. Here are some key features and operations you should be familiar with when working with dictionaries in Python:

 - **Accessing values**: You can access the value of a dictionary item by specifying its key inside square brackets `[]`. For example, `my_dict["key"]` would return the value associated with the key `"key"` in the dictionary `my_dict`.

 - **Modifying values**: Dictionaries are mutable, meaning you can modify their values by assigning new values to specific keys. For example, `my_dict["key"] = "new value"` would set the value of the key `"key"` in the dictionary `my_dict` to `"new value"`.

 - **Adding items**: You can add a new item to a dictionary by specifying a new key-value pair using the syntax `my_dict[new_key] = new_value`. For example, `my_dict["new_key"] = "new_value"` would add a new item to the dictionary `my_dict`.

 - **Removing items**: You can remove an item from a dictionary using the del statement followed by the key of the item you want to remove. For example, `del my_dict["key"]` would remove the item with the key `"key"` from the dictionary `my_dict`.

 - **Length**: You can find the number of items in a dictionary using the `len()` function. For example, `len(my_dict)` would return the number of items in the dictionary `my_dict`.

 - **Iterating**: You can iterate over the keys or values of a dictionary using a `for` loop.

```python
# Accessing values
my_dict = {"name": "Jana", "age": 30, "gender": "female"}
print(my_dict["name"])   # Output: 'Jana'
print(my_dict["age"])    # Output: 30

# Modifying values
my_dict["age"] = 40
print(my_dict["age"])    # Output: 40

# Adding items
my_dict["occupation"] = "engineer"
print(my_dict)           # Output: {'name': 'Jana', 'age': 40, 'gender': 'female', 'occupation': 'engineer'}

# Removing items
del my_dict["gender"]
print(my_dict)           # Output: {'name': 'Jana', 'age': 40, 'occupation': 'engineer'}

# Length
print(len(my_dict))      # Output: 3

# Iterating
for key in my_dict:
    print(key, my_dict[key])
```
